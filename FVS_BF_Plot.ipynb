```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook

#x, p name of graph
#y total number of vertices
#u size of feedback vertex set
#v time on bruteforce approach
#q time on parameterised approach
x,y,z,u,v = np.loadtxt('outputFile.txt',delimiter=':',dtype="U25,i4,U25,i4,f8",unpack='True',skiprows=1)
p,q = np.loadtxt('paraOutputFile.txt', delimiter=':',dtype="U25,f8",unpack='True')

fig = plt.figure(figsize=(12, 10), dpi=80)
ax = fig.add_subplot(111, projection='3d')

i = 0
for i in range(len(x)):
    
    ax.scatter(y[i],u[i],v[i], c='r', marker='o', s=50)
    ax.text(y[i],u[i],v[i], '%s' % (str(i)), size=20,color='r')

i = 0
for i in range(len(p)):
    ax.scatter(y[i],u[i],q[i], c='b', marker='x', s=50)
    ax.text(y[i],u[i],q[i], '%s' % (str(i)), size=20,color='b')
    
#ax.plot_wireframe(x,y,z)
ax.set_xlabel("total vertices")
ax.set_ylabel("fvs size")
ax.set_zlabel("time")

plt.show()
```
